### Introducción a la Prueba de Rachas

La **Prueba de Rachas** (también conocida como prueba de Wald-Wolfowitz) es una prueba estadística no paramétrica que se utiliza para evaluar si una secuencia de datos o eventos se ha generado de forma **aleatoria**.

En este contexto, una **"racha"** se define como una secuencia consecutiva de un mismo evento o característica, precedida y seguida por un evento diferente (o por los límites de la secuencia). Por ejemplo, si lanzamos una moneda y anotamos 0 (Cara) y 1 (Cruz), en la secuencia `0-0-1-0-1-1-1` tenemos 4 rachas:
1. `0-0`
2. `1`
3. `0`
4. `1-1-1`

Esta prueba es muy útil en el control de calidad, el análisis de series temporales y el estudio de residuos en modelos de regresión, ya que nos permite detectar si hay un patrón oculto, una tendencia o un agrupamiento en los datos en lugar de simple ruido aleatorio.

### Hipótesis de la Prueba de Rachas

En el planteamiento de esta prueba, evaluamos si el número total de rachas observadas ($C_0$) es consistente con lo que esperaríamos de un proceso puramente aleatorio. Las hipótesis se plantean formalmente de la siguiente manera:

$$H_0: \text{Los numeros de la muestra {$x_i$} son independientes.}$$

$$H_1: \text{Los numeros de la muestra {$x_i$} no son independientes.}$$

Para muestras grandes (n siendo el numero de la mustra), el número de rachas $C_0$ se aproxima a una distribución normal. En ese caso, utilizamos las siguientes fórmulas para encontrar el estadístico de prueba $Z$:

**Media esperada de rachas:**
$$\mu_{C_0} = \frac{2n - 1}{3}$$

**Varianza esperada:**
$${\sigma_{C_0}}^2 = \frac{16 n - 29}{90}$$

**Estadístico de prueba:**
$${Z_0} = \frac{{C_0} - \mu_{C_0}}{\sigma_{C_0}}$$

*Donde:*
* $C_0$ es el número total de rachas observadas.


Si el valor absoluto de $Z$ es mayor que el valor crítico de la tabla normal estándar para nuestro nivel de significancia (por ejemplo, $|Z_{\frac{\alpha}{2}}| > 1.96$ para un nivel de confianza del 95%), rechazamos la hipótesis nula ($H_0$) y concluimos que los datos no son aleatorios.

```python
# cargar datos
numeros = numeros = [
    0.78961, 0.05230, 0.10699, 0.55877, 0.14151,
    0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989,
    0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121,
    0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850,
    0.79982, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
    0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272,
    0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187,
    0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954,
    0.58244, 0.85853, 0.88752, 0.33729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49383, 0.75103,
    0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724,
    0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]
```

```python


n = len(numeros)
print(f"Total de números procesados (n): {n}\n")

# --- PASO 1: Generar la secuencia de signos (+ y -) ---
signos = []
for i in range(1, n):
    if numeros[i] > numeros[i-1]:
        signos.append('+')
    else:
        signos.append('-')

print("Secuencia S generada:")
print(" ".join(signos))
print("-" * 50)

# --- PASO 2: Contar el número de rachas (C0) ---
C0 = 1  # Iniciamos asumiendo al menos 1 racha
for i in range(1, len(signos)):
    # Si el signo actual es diferente al anterior, es una nueva racha
    if signos[i] != signos[i-1]:
        C0 += 1

print(f"Número de rachas observadas (C0): {C0}")

# --- PASO 3: Calcular Media, Varianza y el Estadístico Z0 ---
# Fórmulas de Rachas Arriba y Abajo
media_C0 = (2 * n - 1) / 3
varianza_C0 = (16 * n - 29) / 90
desviacion_C0 = math.sqrt(varianza_C0)

# Z0 = (C0 - media_C0) / sigma_C0
Z0 = (C0 - media_C0) / desviacion_C0

print(f"Media esperada (mu_C0): {media_C0:.4f}")
print(f"Varianza (varianza_C0): {varianza_C0:.4f}")
print(f"Desviación estándar (desviacion_C0): {desviacion_C0:.4f}")
print(f"Estadístico de prueba (Z0): {Z0:.4f}")
print("-" * 50)

# --- PASO 4: Evaluación y Conclusión ---
# Para un nivel de significancia alfa = 0.05, Z_alfa/2 es 1.96
Z_critico = 1.96
print(f"Intervalo de aceptación: [-{Z_critico}, {Z_critico}]")

if -Z_critico <= Z0 <= Z_critico:
    print("\n CONCLUSIÓN: El valor Z0 cae DENTRO del intervalo.")
    print("NO SE RECHAZA H0. Los números de la muestra SON INDEPENDIENTES.")
else:
    print("\n CONCLUSIÓN: El valor Z0 cae FUERA del intervalo.")
    print("SE RECHAZA H0. Los números de la muestra NO SON INDEPENDIENTES.")
```
